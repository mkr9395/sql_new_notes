**1. Question:** Explain the difference between `WHERE` and `HAVING` clauses in SQL. When would you use each?

**Answer:**
   - `WHERE` filters rows *before* aggregation, operating on individual rows.
   - `HAVING` filters rows *after* aggregation, operating on grouped rows.
   - Use `WHERE` for row-level conditions and `HAVING` for conditions based on aggregated values (e.g., `SUM`, `AVG`, `COUNT`).

**2. Question:** Describe the purpose of window functions in SQL and provide an example using `RANK()` or `DENSE_RANK()`.

**Answer:**
   - Window functions perform calculations across a set of rows related to the current row without collapsing them into a single output row.
   - Example (using `DENSE_RANK()`):
     ```sql
     SELECT product_id, sales, DENSE_RANK() OVER (ORDER BY sales DESC) AS sales_rank
     FROM sales_table;
     ```
     This ranks products by sales, handling ties without gaps.

**3. Question:** How do you optimize a slow-running SQL query? List at least three techniques.

**Answer:**
   - **Indexing:** Ensure appropriate indexes are created on frequently queried columns.
   - **Query Rewriting:** Avoid `SELECT *`, use specific column names, and minimize subqueries.
   - **Avoid Correlated Subqueries:** Replace them with `JOIN` operations or window functions.
   - **Partitioning:** Partition large tables to improve query performance on subsets of data.
   - **Execution plan analysis:** Use `EXPLAIN` to understand how the query is being executed and identify bottlenecks.

**4. Question:** Explain the concept of a "common table expression" (CTE) and provide a use case.

**Answer:**
   - A CTE is a temporary named result set that you can reference within a `SELECT`, `INSERT`, `UPDATE`, or `DELETE` statement.
   - Use case: Breaking down complex queries into smaller, more readable parts or recursive queries.
     ```sql
     WITH MonthlySales AS (
         SELECT DATE_TRUNC('month', order_date) AS month, SUM(sales) AS total_sales
         FROM orders
         GROUP BY month
     )
     SELECT month, total_sales
     FROM MonthlySales
     WHERE total_sales > 10000;
     ```

**7. Question:** Explain the difference between `INNER JOIN`, `LEFT JOIN`, `RIGHT JOIN`, and `FULL OUTER JOIN`.

**Answer:**
   - `INNER JOIN`: Returns rows where there's a match in both tables.
   - `LEFT JOIN`: Returns all rows from the left table and matched rows from the right table (NULLs for unmatched rows).
   - `RIGHT JOIN`: Returns all rows from the right table and matched rows from the left table (NULLs for unmatched rows).
   - `FULL OUTER JOIN`: Returns all rows when there's a match in either table.

**8. Question:** Describe how to implement a recursive query in SQL.

**Answer:**
   - Use a CTE with a `UNION ALL` clause, where the CTE references itself.
   - This is useful for hierarchical data like organizational structures or bill-of-materials.
     ```sql
     WITH RECURSIVE EmployeeHierarchy AS (
         SELECT employee_id, manager_id, employee_name, 0 AS level
         FROM employees
         WHERE manager_id IS NULL
         UNION ALL
         SELECT e.employee_id, e.manager_id, e.employee_name, eh.level + 1
         FROM employees e
         JOIN EmployeeHierarchy eh ON e.manager_id = eh.employee_id
     )
     SELECT employee_name, level
     FROM EmployeeHierarchy;
     ```
